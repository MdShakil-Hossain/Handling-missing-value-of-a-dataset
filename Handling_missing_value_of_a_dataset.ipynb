```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy import stats
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
data = pd.read_csv("/content/drive/MyDrive/CSE303/Lab 7/Heart_Attack_Risk_Prediction_Dataset_2.csv")
```

```python
data
```

```python
data = data.drop(columns='Patient ID')
data.head()
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data.shape
```

```python
import numpy as np

# Custom Log Scaled Frequency Encoding
for col in ['Sex', 'Diet', 'Continent']:
    freq_map = data[col].value_counts(normalize=True).to_dict()

    # Apply log transformation to compress high values
    data[col] = data[col].map(lambda x: np.log1p(freq_map.get(x, 0.01)))  # Avoid log(0)
```

```python
data.head(10)
```

```python
from sklearn.impute import SimpleImputer

# Initialize imputer with mean strategy
imputer = SimpleImputer(strategy='mean')

# Apply imputation and update the dataframe
data[:] = imputer.fit_transform(data)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```
